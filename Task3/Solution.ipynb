Отдельно отметим, что модель может ошибаться и выделять в качестве самолетов части изображений, которые самолетов и не содержат. В любом случае, выделенные моделью области будем называть "самолетами".

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
```

```python
# Загрузка изображения
image_path = 'Images/Images/428481.jpg'
image = cv2.imread(image_path)
if image is None:
    raise FileNotFoundError(f"Image {image_path} not found")
```

```python
# Создание объекта для селективного поиска
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
ss.setBaseImage(image)
ss.switchToSelectiveSearchFast()
```

```python
# Выполнение селективного поиска
rects = ss.process()
print(f"Total number of region proposals: {len(rects)}")
```

```python
# Загрузка предобученной модели из локальной директории
model_path = 'Comp_Vision_Task_3_model/ieeercnn_vgg16_1.h5'
model = tf.keras.models.load_model(model_path)
```

- Введите координату y левого верхнего угла соответствующего прямоугольника

```python
from matplotlib import pyplot as plt
# Порог вероятности для определения самолета
threshold = 0.65
imout = image.copy()

# Обнаружение самолетов
airplanes_count = 0
max_probability = 0.0
min_probability = float('inf')
best_rect = None
worst_rect = None
for (x, y, w, h) in rects:
    roi = image[y:y+h, x:x+w]
    if roi.shape[0] == 0 or roi.shape[1] == 0:
        continue
    resized = cv2.resize(roi, (224,224), interpolation = cv2.INTER_AREA)
    img = np.expand_dims(resized, axis = 0)
    predictions= model.predict(img)
    if predictions[0][0] > threshold:
        cv2.rectangle(imout, (x, y), (x + w, y + h), (0, 255, 0), 1, cv2.LINE_AA)
        airplanes_count += 1
        if predictions[0][0] > max_probability:
            max_probability = predictions[0][0]
            best_rect = (x, y, w, h)
        if predictions[0][0] < min_probability:
            min_probability = predictions[0][0]
            worst_rect = (x, y, w, h)
plt.figure()
plt.imshow(imout)

print(f"Number of airplanes detected: {airplanes_count}")
print(f"Maximum probability of airplane detection: {max_probability}")
if best_rect:
    x, y, w, h = best_rect
    print(f"Coordinates of the top-left corner of the rectangle: ({x}, {y})")
    print(f"Width of the rectangle: {w} pixels")
    print(f"Height of the rectangle: {h} pixels")
else:
    print("No airplanes detected with the given threshold.")
print(f"Minimum probability of airplane detection: {min_probability}")
if worst_rect:
    x, y, w, h = worst_rect
    print(f"Coordinates of the top-left corner of the rectangle: ({x}, {y})")
    print(f"Width of the rectangle: {w} pixels")
    print(f"Height of the rectangle: {h} pixels")
else:
    print("No airplanes detected with the given threshold.")
```
